```python
import os, sys

sys.path.append(os.path.abspath("../src"))

from predict_genre import predict_genre

# Use a RELATIVE path instead of absolute
my_audio = os.path.abspath("../test_audio/Bring It Together - Telecasted.mp3")

pred, proba = predict_genre(my_audio)
print("Predicted genre:", pred)

if proba is not None:
    for g, p in sorted(proba.items(), key=lambda kv: kv[1], reverse=True):
        print(f"{g:10s} : {p:.3f}")
```

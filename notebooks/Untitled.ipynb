```python
from pptx import Presentation
from pptx.util import Inches

prs = Presentation()

# ---------- helpers ----------

def add_title_content(title, bullet_lines):
    """Slide with title + bullet list."""
    slide_layout = prs.slide_layouts[1]  # Title and Content
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.title.text = title

    body = slide.placeholders[1]
    tf = body.text_frame
    tf.clear()

    for i, line in enumerate(bullet_lines):
        if i == 0:
            p = tf.paragraphs[0]
        else:
            p = tf.add_paragraph()
        p.text = line
        p.level = 0


def add_image_slide(title, img_path, caption=None):
    """Slide with title + big image + optional caption."""
    slide_layout = prs.slide_layouts[5]  # Title Only
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.title.text = title

    # image
    slide.shapes.add_picture(img_path, Inches(1), Inches(1.4), width=Inches(8))

    # caption (small text box at bottom)
    if caption:
        left, top, width, height = Inches(1), Inches(6.5), Inches(8), Inches(1)
        txBox = slide.shapes.add_textbox(left, top, width, height)
        tf = txBox.text_frame
        tf.text = caption


# ---------- slides ----------

# 1. Title
slide = prs.slides.add_slide(prs.slide_layouts[0])
slide.shapes.title.text = "Music Genre Classification"
slide.placeholders[1].text = "Hamiz Siddiqui"

# 2. Overview / Motivation
add_title_content(
    "Project Overview",
    [
        "Goal: automatically classify songs into genres using machine learning.",
        "Motivation: similar idea used in Spotify/Apple Music for recommendations and playlist generation.",
        "Focus: classical ML models with engineered audio features instead of deep learning."
    ]
)

# 3. Dataset
add_title_content(
    "Dataset: GTZAN",
    [
        "999 usable audio clips across 10 genres (blues, classical, country, disco, hiphop, jazz, metal, pop, reggae, rock).",
        "Each clip is ~30 seconds, sampled at 22,050 Hz.",
        "One corrupted file was removed; train/test split = 80% / 20% with stratification."
    ]
)

# 4. Feature Extraction
add_title_content(
    "Feature Extraction (Librosa)",
    [
        "Extracted MFCCs, chroma vectors, spectral contrast, tonnetz, spectral centroid/bandwidth/rolloff, and RMS energy.",
        "Computed mean and standard deviation over time for each feature type.",
        "Final representation: 157-dimensional feature vector per track."
    ]
)

# 5. Models and Training
add_title_content(
    "Models and Training",
    [
        "Supervised models: Logistic Regression, kNN, Linear SVM, RBF SVM, Decision Tree, Random Forest, AdaBoost.",
        "Used GridSearchCV with 5-fold cross-validation to tune hyperparameters.",
        "Also tried unsupervised clustering (K-Means, Bayesian GMM) and generative models (GaussianNB, class-conditional GMMs)."
    ]
)

# 6. Model Comparison table
add_image_slide(
    "Model Performance Comparison",
    "model_table.png",
    caption="Random Forest achieved the best test accuracy (~78.5%) and macro F1-score (~0.78)."
)

# 7. Learning Curve
add_image_slide(
    "Learning Curve",
    "learning_curve.png",
    caption="Validation accuracy improves as we add more training examples; training accuracy remains very high, indicating some overfitting but room for improvement with more data."
)

# 8. Validation Curve
add_image_slide(
    "Validation Curve (SVM C parameter)",
    "validation_curve.png",
    caption="Best performance occurs around C ≈ 1–10; low C underfits, very large C starts to overfit."
)

# 9. Confusion Matrix
add_image_slide(
    "Confusion Matrix (Random Forest)",
    "confusion_matrix.png",
    caption="Classical, metal, and jazz are classified very accurately; country, rock, and reggae are more often confused with each other."
)

# 10. PCA
add_image_slide(
    "PCA Visualization",
    "pca.png",
    caption="PCA projection shows heavy overlap between genres in feature space, explaining why unsupervised clustering is difficult."
)

# 11. t-SNE
add_image_slide(
    "t-SNE Visualization",
    "tsne.png",
    caption="t-SNE reveals some curved manifold structure but genres are still mixed, confirming that labels are important for this task."
)

# 12. Insights and Future Work
add_title_content(
    "Insights and Future Work",
    [
        "Engineered audio features plus classical ML (especially Random Forest) can reach ~78.5% test accuracy.",
        "Genres with distinct timbres (classical, metal, jazz) are easier; overlapping styles (rock/country/reggae) are harder.",
        "Future work: CNNs on mel-spectrograms, data augmentation, larger datasets, and sequence models such as CRNNs or transformers."
    ]
)

# 13. Conclusion
add_title_content(
    "Conclusion",
    [
        "Built a complete pipeline: feature extraction, model training, evaluation, and a script that predicts genre for a new MP3 file.",
        "Results show classical ML is still competitive for music genre classification with good features.",
        "Project gave hands-on experience with audio processing, model comparison, and interpreting ML results."
    ]
)

# ---------- save ----------
output_path = "music_genre_classification_presentation_full.pptx"
prs.save(output_path)
print("Saved presentation to:", output_path)
```
